````python
#write a program to print 1 to 50 numbers using for loop?
for num in range(1,51):
    print(num)
````

````python
#write a program to iterate name using for loop
name="shivaji"
for char in name:
    print(char)
````

````python
#write a program to find the even numbers in between 100 to 150 using for loop?
for n in range(100,151):
    if(n%2==0):
        print(n,end=",")
````

````python
#write a program tpo find the odd numbers in between 100 to 150 using for loop?
for n in range(100,151):
    if(n%2==1):
       print(n,end=",")
````

````python
#write a program to find the vowles from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if(char in "aeiouAEIOU"):
      print(char)
````

````python
#write a program to find non-vowels from string using for loop?
name=str(input("please enter name here:"))
for char in name:
     if(char not in "aeiouAEIOU"):
         print(char)
````

````python
#write a program to fine list using for loop
names=["vikas","shivaji","aditya","chandra"]
for n in names:
    print('my name is',n)
````

````python
#write a program ton find sum of all numbers in between 1 to 100 using for loop?
total = 0
for n in range(1, 101):
    total += n
print("Sum of numbers from 1 to 100 is:", total)
````

````python
#write a program to find the sum of all even numbers in between 1 to 50 using for loop?
total = 0
for n in range(2,51,2):
 total += n
print("sum of even numbers from 1 to 50 is:",total)
````

````python
#write a program to find the sum of all odd numbers in between 1 to 50 using for loop?
total = 0
for n in range(1,51,1):
    total += n
print("sum of odd numbers from 1 to 50 is:",total)
````
